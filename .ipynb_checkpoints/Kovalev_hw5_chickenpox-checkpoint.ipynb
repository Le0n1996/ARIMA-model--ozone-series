```python
import pandas as pd
from scipy import stats
import statsmodels
import statsmodels.api as sm
import matplotlib.pyplot as plt
import warnings
import numpy as np
from pandas.tseries.offsets import MonthBegin
from itertools import product

plt.style.use('ggplot')
    
%matplotlib inline

plt.rcParams['figure.figsize'] = (15, 8)
```

В рамках данного задания необходимо было выбрать любой нетривиальный (не константный и т.п.) временной ряд длиной от 100 отсчетов, проанализировать его по заданной схеме (как на семинаре) и построить прогноз на 3 года вперед с помощью модели класса ARIMA. Я взял временной ряд, показывающий количество зафиксированных случаев ветрянки в Нью-Йорке с 1931 по 1972 гг. с дискретизацией в месяц. Он доступен по ссылке:
https://datamarket.com/data/set/22v7/monthly-reported-number-of-chickenpox-new-york-city-1931-1972

# 1 Временной ряд и его график

```python
data = pd.read_csv('monthly-reported-number-of-chickenpox.csv', index_col=['month'],
                   parse_dates=['month'])
data.head()
```

```python
data.tail()
```

```python
data['chickenpox'].plot()
plt.ylabel('Number of chickenpox')
plt.show()
```

# 2 STL

```python
sm.tsa.seasonal_decompose(data['chickenpox']).plot()
plt.show()
```

```python
print("Критерий Дики-Фуллера: p = %f" % sm.tsa.stattools.adfuller(data['chickenpox'])[1])
```

# 3 Стабилизация дисперсии

Из графика видно, что дисперсия нестабильна - разберемся с этим при помощи преобразования Бокса-Кокса.

```python
data['chickenpox_box'], lmbda = stats.boxcox(data['chickenpox'])
data['chickenpox_box'].plot()
plt.ylabel('Transformed number of chickenpox')
print("Оптимальный параметр преобразования Бокса-Кокса: %f" % lmbda)
```

Как видно из графика, дисперсия стабилизировалась.

# 4 Стационарность

```python
print("Критерий Дики-Фуллера: p = %f" % sm.tsa.stattools.adfuller(data['chickenpox_box'])[1])
```

Итак, гипотеза о нестационарности не отвергается. Сделаем сезонное дифференцирование:

```python
data['chickenpox_box_diff'] = data['chickenpox_box'] - data['chickenpox_box'].shift(12)
fig = sm.tsa.seasonal_decompose(data['chickenpox_box_diff'][12:]).plot()
print("Критерий Дики-Фуллера: p = %f" % sm.tsa.stattools.adfuller(data['chickenpox_box_diff'][12:])[1])
```

Видно, что от тренда мы избавились, и гипотеза о нестационарности отвергается в соответствии с критерием Дики-Фуллера.

# 5 ACF/PACF

```python
fig = sm.graphics.tsa.plot_acf(data['chickenpox_box_diff'][12:].values.squeeze(), lags=72)
```

```python
fig = sm.graphics.tsa.plot_pacf(data['chickenpox_box_diff'][12:].values.squeeze(), lags=72)
```

Из данных коррелограмм можно получить начальные приближения необходимых параметров предсказывающей модели.

Из графика ACF:

- Q - номер последнего сезонного лага, при котором автокорреляция значима (2)

- q - номер последнего несезонного лага, при котором автокорреляция значима (5)

Из графика PACF:

- P - номер последнего сезонного лага, при котором частичная автокорреляция значима (4)

- p - номер последнего несезонного лага, при котором частичная автокорреляция значима (1)

# 6 Выбор лучшей модели

```python
# D - сколько раз применялось сезонное дифференцирование
D = 1
# d - сколько раз применялось обычное дифференцирование
d = 0
ps = range(0, 2)
qs = range(0, 6)
Ps = range(0, 5)
Qs = range(0, 3)
```

```python
parameters = product(ps, qs, Ps, Qs)
parameters_list = list(parameters)
len(parameters_list)
```

Теперь, перебирая модели с разными параметрами p, q, P, Q и сравнивая их по критерию Акаике, найдем лучшую.

```python
%%time
results = []
best_aic = float("inf")
warnings.filterwarnings('ignore')

for param in parameters_list:
    #try except нужен, потому что на некоторых наборах параметров модель не обучается
    try:
        model = sm.tsa.statespace.SARIMAX(data['chickenpox_box'], order=(param[0], d, param[1]), 
                                          seasonal_order=(param[2], D, param[3], 12)).fit(disp=-1)
    # выводим параметры, на которых модель не обучается и переходим к следующему набору
    except ValueError:
        print('wrong parameters:', param)
        continue
    aic = model.aic
    # сохраняем лучшую модель, aic, параметры
    if aic < best_aic:
        best_model = model
        best_aic = aic
        best_param = param
    results.append([param, model.aic])
    
warnings.filterwarnings('default')
```

```python
result_table = pd.DataFrame(results)
result_table.columns = ['parameters', 'aic']
print(result_table.sort_values(by = 'aic', ascending=True).head())
```

Лучшая модель:

```python
print(best_model.summary())
```

# 7 Остатки

```python
plt.subplot(211)
best_model.resid[12:].plot()
plt.ylabel(u'Residuals')

ax = plt.subplot(212)
sm.graphics.tsa.plot_acf(best_model.resid[12:].values.squeeze(), lags=48, ax=ax)

print("Критерий Стьюдента: p = %f" % stats.ttest_1samp(best_model.resid[12:], 0)[1])
print("Критерий Дики-Фуллера: p = %f" % sm.tsa.stattools.adfuller(best_model.resid[12:])[1])
```

Свойства остатков:

- несмещены (по критерию Стьюдента)

- стационарны (видно из графика остатков, но если строго, то по критерию Дики-Фуллера)

- несмотря на то, что критерий Льюнга-Бокса указывает на возможный неучет всех особенностей данных (то есть, наверное, можно придумать модель чуть получше), из коррелограммы видно, что остатки в целом неавтокоррелированы. Поэтому я решил не особо заморачиваться насчет идеального качества остатков и остановиться на данной модели.

# 8 Качество модели на исходных данных

```python
# обратное преобразование Бокса-Кокса
def invboxcox(y, lmbda):
    if lmbda == 0:
        return (np.exp(y))
    else:
        return(np.exp(np.log(y * lmbda + 1) / lmbda))

data['model'] = invboxcox(best_model.fittedvalues, lmbda)
data['chickenpox'].plot(color='r')
data['model'][12:].plot(color='b')
plt.ylabel('Number of chickenpox')
plt.show()
```

Качество модели хорошее, но видно, что в местах больших скачков прогноз немного "не дотягивает" до реального значения. И в принципе заметно, что дисперсия прогноза в целом чуть меньше, чем дисперсия реальных значений. Возможно, это связано с тем, что не было произведено обычного дифференцирования.

# 9 Прогноз

Чтобы лучше было видно предсказание на фоне исходных данных, я решил спрогнозировать не на 3 года вперед, а на 10. Какая, собственно, разница, что 3, что 10...

```python
data2 = data[['chickenpox']]
date_list = pd.date_range("1972-06-01", periods=120, freq='m') + MonthBegin(n=0)
future = pd.DataFrame(index=date_list, columns=data2.columns)
data2 = pd.concat([data2, future])
data2['forecast'] = invboxcox(best_model.predict(start=498, end=617), lmbda)

data2['chickenpox'].plot(color='r')
data2['forecast'].plot(color='b')
plt.ylabel('Number of chickenpox')
plt.show()
```

```python
data2.drop(['chickenpox'], axis=1)[498:].to_csv("chickenpox_forecast.csv")
```
