- выбрать любой временной ряд, который вы будете прогнозировать; лучше взять ряд с месячным или квартальным разрешением, длиной от 100 отсчётов и не слишком простой (не выглядящий как константа или другая простая функция), иначе вам будет скучно его прогнозировать;
- внести выбранный ряд в таблицу, убедившись, что никто другой его не взял - если несколько студентов будут прогнозировать один и тот же ряд, задача не будет засчитана никому из них;
- проанализировать выбранный ряд по аналогии с примером, разбиравшимся на семинаре, и построить его прогноз на 3 года вперёд с помощью модели класса ARIMA.

```python
import pandas as pd
import numpy as np
from scipy import stats
import statsmodels
import statsmodels.api as sm
import matplotlib.pyplot as plt
import warnings
from itertools import product
from pandas.tseries.offsets import MonthBegin

plt.style.use('bmh')
%matplotlib inline 
plt.rcParams['figure.figsize'] = (15, 7)

def invboxcox(y, lmbda):
   if lmbda == 0:
      return(np.exp(y))
   else:
      return(np.exp(np.log(lmbda*y+1)/lmbda))
```

```python
df = pd.read_csv('ozon-concentration-downtown-l-a-.csv', index_col=['Month'], parse_dates=['Month'])
df.head()
```

```python
df.tail()
```

```python
df['ozone'].plot()
plt.ylabel('Ozone concentration')
plt.show()
```

### Проверка стационарности и STL-декомпозиция ряда:

Проверяем, насколько наш временной ряд не стационарен (хотя и из графика видно, что стационарности никакой нет):

```python
sm.tsa.seasonal_decompose(df['ozone']).plot()
plt.show()
```

Применяем критерий Дики-Фуллера:

```python
print("Критерий Дики-Фуллера: p = %f" % sm.tsa.stattools.adfuller(df['ozone'])[1])
```

Т.е. стационарности нет.

### Стабилизируем дисперсию при помощи преобразования Бокса-Кокса:

```python
df['ozone_boxcox'], opt = stats.boxcox(df['ozone'])

df['ozone_boxcox'].plot()
plt.ylabel('Ozone concentration (transformed)')
print("Оптимальный параметр преобразования Бокса-Кокса: %f" % opt)
print("Критерий Дики-Фуллера: p = %f" % sm.tsa.stattools.adfuller(df['ozone_boxcox'])[1])
```

Дисперсия, как можно видеть на графике, совсем немного выравнялась.

### Стационарность:

По критерию Дики-Фуллера гипотеза стационарности отвергается (p > 0.01), в данных виден тренд. Выделим его, применив сезонное дифференцирование, после чего для продифференцированного ряда также делаем STL-декомпозицию и проверяем его на стационарность:

```python
df['ozone_boxcox_diff'] = df['ozone_boxcox'] - df['ozone_boxcox'].shift(12)
sm.tsa.seasonal_decompose(df['ozone_boxcox_diff'][12:]).plot()
print("Критерий Дики-Фуллера: p = %f" % sm.tsa.stattools.adfuller(df['ozone_boxcox_diff'][12:])[1])
```

Видно, что теперь p < 0.01, стабильного тренда почти нет, и гипотеза о нестационарности нашего ряда отвергается (с высоким уровнем значимости). Попробуем еще чуть-чуть улучшить ситуацию, добавив обычное дифференцирование, а не сезонное:

```python
df['ozone_boxcox_diff2'] = df.ozone_boxcox_diff - df.ozone_boxcox_diff.shift(1)
sm.tsa.seasonal_decompose(df.ozone_boxcox_diff2[13:]).plot()   
print("Критерий Дики-Фуллера: p = %f" % sm.tsa.stattools.adfuller(df.ozone_boxcox_diff2[13:])[1])
```

Теперь гипотеза о нестационарности отвергается с уровнем значимости более 0,99, тренда нет.

### Подбор модели:

Смотрим на ACF и PACF:

```python
ax = plt.subplot(211)
sm.graphics.tsa.plot_acf(df.ozone_boxcox_diff2[13:].values.squeeze(), lags=60, ax=ax)
plt.show()
ax = plt.subplot(212)
sm.graphics.tsa.plot_pacf(df.ozone_boxcox_diff2[13:].values.squeeze(), lags=60, ax=ax)
plt.show()
```

- Q = 1, потому что на графике ACF наблюдается значимое значение на лаге 12
- q = 2, потому что на графике ACF налюдается значимое значение на лагах 1-2
- P = 1, потому что на графике PACF наблюдается значимое значение на лаге 12
- p = 3, потому что на графике PACF налюдается значимое значение на лагах 1-3

Поэтому начальные приближения - Q = 1, q = 2, P = 1, p = 3.

```python
D=1 # мы один раз использовали сезонное дифференцирование
d=1 # мы один раз использовали простое дифференцирование

ps = range(0, 4) # последнее значение не входит!
qs = range(0, 3)
Ps = range(0, 2)
Qs = range(0, 2)

params = product(ps, qs, Ps, Qs)
params_list = list(params)
```

```python
%%time
results = []
best_aic = float("inf")
warnings.filterwarnings('ignore')

for param in params_list:
    #try except нужен, потому что на некоторых наборах параметров модель не обучается
    try:
        model=sm.tsa.statespace.SARIMAX(df.ozone_boxcox, order=(param[0], d, param[1]), 
                                        seasonal_order=(param[2], D, param[3], 12)).fit(disp=-1)
    #выводим параметры, на которых модель не обучается и переходим к следующему набору
    except ValueError:
        print('wrong parameters:', param)
        continue
    aic = model.aic # используем информационный критерий Акаике
    #сохраняем лучшую модель, aic, параметры
    if aic < best_aic:
        best_model = model
        best_aic = aic
        best_param = param
    results.append([param, model.aic])
    
warnings.filterwarnings('default')
```

```python
result_table = pd.DataFrame(results)
result_table.columns = ['parameters', 'aic']
print(result_table.sort_values(by = 'aic', ascending=True).head()) # Чем меньше aic, тем лучше модель, сравниваем:
```

```python
print(best_model.summary())
```

### Остатки:

```python
plt.subplot(211)
best_model.resid[13:].plot()
plt.ylabel('Residuals')

ax = plt.subplot(212)
sm.graphics.tsa.plot_acf(best_model.resid[13:].values.squeeze(), lags=60, ax=ax)

print("Критерий Стьюдента: p = %f" % stats.ttest_1samp(best_model.resid[13:], 0)[1])
print("Критерий Дики-Фуллера: p = %f" % sm.tsa.stattools.adfuller(best_model.resid[13:])[1])
```

Для остатков выполнено следующее:
- средний остаток = 0, т.е. остатки несмещенные (критерий Стьюдента подтверждает это),
- остатки стационарны (видно на графике, подтверждается критерием Дики-Фуллера для них),
- остатки неавтокоррелированы (подтверждается критерием Льюнга-Бокса).

### Посмотрим, насколько хорошо модель описывает исходные данные:

```python
df['model'] = invboxcox(best_model.fittedvalues, opt)
df.ozone.plot()
df.model[13:].plot(color='r')
plt.ylabel('Ozone concentration')
plt.show()
```

Получилось неплохо, наша модель очень хорошо описывает исходные данные. Можно составлять прогноз.

### Прогноз на три года вперед:

```python
ozone_exp = df[['ozone']]
date_list = pd.date_range("1972-12-01", periods=36, freq='M') + MonthBegin(n=0)
future = pd.DataFrame(index=date_list, columns= ozone_exp.columns)
ozone_exp = pd.concat([ozone_exp, future])
ozone_exp['forecast'] = invboxcox(best_model.predict(start=216, end=251), opt)

ozone_exp.ozone.plot()
ozone_exp.forecast.plot(color='r')
plt.ylabel('Ozone concentration')
plt.show()
```

Прогноз построен, его можно выгрузить в виде csv-файла с помощью кода ниже (комментарий предотвращает выгрузку при запуске)

```python
ozone_exp.drop(['ozone'], axis=1)[216:].to_csv("ozone_forecast.csv")
```
